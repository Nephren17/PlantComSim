```python
import numpy as np

def simulate_tropical_savanna_weather():


    # 平均气温
    temp = []
    for i in range(12):
        # 亚热带稀树草原的平均温度为18-22摄氏度
        temp_month = np.random.normal(loc=20, scale=8)
        if i in [0,1,11]: # 冬季
            temp_month = np.random.normal(loc=10, scale=8)
        elif i in [6,7,8]: # 夏季
            temp_month = np.random.normal(loc=26, scale=8)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 亚热带稀树草原的平均湿度为50-70%
        humidity_month = np.random.normal(loc=60, scale=10)
        if i in [5,6,7]: # 雨季
            humidity_month = np.random.normal(loc=80, scale=10)
        elif i in [1,2,11]: # 旱季
            humidity_month = np.random.normal(loc=40, scale=10)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 亚热带稀树草原的平均降水量为100毫米
        precipitation_month = np.random.normal(loc=100, scale=50)
        if i in [5,6,7]: # 雨季
            precipitation_month = np.random.normal(loc=200, scale=100)
        elif i in [1,2,11]: # 旱季
            precipitation_month = np.random.normal(loc=50, scale=20)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 亚热带稀树草原的平均日照时数为2200-2600小时
        sunshine_month = np.random.normal(loc=2400, scale=200)
        if i in [0,1,11]: # 冬季
            sunshine_month = np.random.normal(loc=2000, scale=200)
        elif i in [6,7,8]: # 夏季
            sunshine_month = np.random.normal(loc=2800, scale=200)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
def simulate_tropical_rainforest_weather():

    # 平均气温
    temp = []
    for i in range(12):
        # 热带雨林气候平均温度是25-27摄氏度
        temp_month = np.random.normal(loc=25, scale=2)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 热带雨林气候的平均湿度可达80%至90%
        humidity_month = np.random.normal(loc=85, scale=8)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 热带雨林气候的平均降水量为200毫米
        precipitation_month = np.random.normal(loc=200, scale=50)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 热带雨林气候的平均光照量为1500小时
        sunshine_month = np.random.normal(loc=1500, scale=200)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
def simulate_tropical_dessert_weather():

    # 平均气温
    temp = []
    for i in range(12):
        # 热带沙漠气候平均温度是20-30摄氏度
        temp_month = np.random.normal(loc=25, scale=5)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 热带沙漠气候的平均湿度可达10%至25%
        humidity_month = np.random.normal(loc=15, scale=8)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 热带沙漠气候的平均降水量5毫米
        precipitation_month = np.random.normal(loc=5, scale=3)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 热带沙漠气候的平均光照量为2700小时
        sunshine_month = np.random.normal(loc=2700, scale=400)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
import numpy as np

def simulate_temperate_grassland_climate(random_seed=None):
    # 设置随机数生成器的种子
    np.random.seed(random_seed)
    
    # 生成每个月的平均气温
    temperature = []
    for i in range(12):
        # 夏季的平均气温较高，冬季的平均气温较低
        if i in [0, 1, 11]:
            t = np.random.normal(5, 2)
        elif i in [2, 3, 4, 9, 10]:
            t = np.random.normal(15, 2)
        else:
            t = np.random.normal(-5, 2)
        temperature.append(t)
    
    # 生成每个月的降水量
    precipitation = []
    for i in range(12):
        # 夏季的降水量较多，冬季的降水量较少
        if i in [5, 6, 7]:
            p = np.random.normal(100, 20)
        elif i in [0, 1, 11]:
            p = np.random.normal(40, 10)
        else:
            p = np.random.normal(20, 5)
        precipitation.append(p)
    
    # 生成每个月的湿度
    humidity = []
    for i in range(12):
        # 湿度与降水量有关，夏季湿度较高，冬季湿度较低
        if i in [5, 6, 7]:
            h = min(100,np.random.normal(55, 5))
        elif i in [0, 1, 11]:
            h = np.random.normal(45, 5)
        else:
            h = np.random.normal(35, 5)
        humidity.append(h)
    
    # 生成每个月的光照量
    sunshine = []
    for i in range(12):
        # 光照量与降水量和云量有关，夏季光照量较高，冬季光照量较低
        if i in [5, 6, 7]:
            s = np.random.normal(240, 20)
        elif i in [0, 1, 11]:
            s = np.random.normal(160, 20)
        else:
            s = np.random.normal(180, 20)
        sunshine.append(s)
    
    return temperature, humidity, precipitation, sunshine
```

```python
import matplotlib.pyplot as plt

def plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight):
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
    months = range(1, 13)

    # Plot temperature
    axs[0, 0].plot(months, year_temp, marker='o')
    axs[0, 0].set_title('Average Monthly Temperature (Celsius)')
    axs[0, 0].set_xlabel('Month')
    axs[0, 0].set_xticks(months)

    # Plot humidity
    axs[0, 1].plot(months, year_humidity, marker='o')
    axs[0, 1].set_title('Average Monthly Humidity (%)')
    axs[0, 1].set_xlabel('Month')
    axs[0, 1].set_xticks(months)

    # Plot rainfall
    axs[1, 0].plot(months, year_rainfall, marker='o')
    axs[1, 0].set_title('Average Monthly Rainfall (mm)')
    axs[1, 0].set_xlabel('Month')
    axs[1, 0].set_xticks(months)

    # Plot sunlight
    axs[1, 1].plot(months, year_sunlight, marker='o')
    axs[1, 1].set_title('Average Monthly Sunlight (hours)')
    axs[1, 1].set_xlabel('Month')
    axs[1, 1].set_xticks(months)

    # Adjust spacing and show plot
    fig.tight_layout()
    plt.show()
```

```python
year_temp, year_humidity, year_rainfall, year_sunlight = simulate_tropical_savanna_weather()

plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight)
```

```python
def simulate_tropical_savanna_weather_daily():


    # 平均气温
    temp = []
    for i in range(360):
        # 亚热带稀树草原的平均温度为18-22摄氏度
        temp_month = np.random.normal(loc=20, scale=8)
        if i//30 in [0,1,11]: # 冬季
            temp_month = np.random.normal(loc=10, scale=8)
        elif i//30  in [6,7,8]: # 夏季
            temp_month = np.random.normal(loc=26, scale=8)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(360):
        # 亚热带稀树草原的平均湿度为50-70%
        humidity_month = np.random.normal(loc=60, scale=10)
        if i//30  in [5,6,7]: # 雨季
            humidity_month = np.random.normal(loc=80, scale=10)
        elif i//30  in [1,2,11]: # 旱季
            humidity_month = np.random.normal(loc=40, scale=10)
        humidity.append(humidity_month)

    # 平均昼夜温差
    diurnal = []
    for i in range(360):
        # 亚热带稀树草原的平均温度为10-30摄氏度
        diurnal_month = np.random.normal(loc=20, scale=10)
        diurnal.append(diurnal_month)


    return temp, humidity, diurnal
```

```python
import numpy as np

def simulate_tropical_savanna_climate_daily():
    # 设置参数
    num_days = 360
    temp_range_rainy = 8
    temp_range_dry = 15
    temp_mean_rainy = 27
    temp_mean_dry = 22
    temp_amplitude = 5
    rain_amplitude = 30
    dry_start_day = 91  # 4月1日是第91天
    dry_end_day = 273  # 9月30日是第273天
    
    # 计算每天的平均气温和昼夜温差
    temp = np.zeros(num_days)
    diurnal_temp_range = np.zeros(num_days)
    for i in range(num_days):
        if i >= dry_start_day and i <= dry_end_day:  # 旱季
            temp[i] = np.random.normal(temp_mean_dry, temp_range_dry / 2)
            diurnal_temp_range[i] = np.random.normal(temp_amplitude, temp_amplitude / 4)
        else:  # 雨季
            temp[i] = np.random.normal(temp_mean_rainy, temp_range_rainy / 2)
            diurnal_temp_range[i] = np.random.normal(temp_amplitude / 2, temp_amplitude / 8)
            
    # 计算每天的降水量
    rain = np.zeros(num_days)
    for i in range(num_days):
        if i >= dry_start_day and i <= dry_end_day:  # 旱季
            rain[i] = np.random.normal(0, rain_amplitude / 2)
        else:  # 雨季
            rain[i] = np.random.normal(rain_amplitude, rain_amplitude / 4)
            if rain[i] < 0:
                rain[i] = 0
                
    return temp, rain, diurnal_temp_range

```

```python
import numpy as np

def simulate_temperate_grassland_climate_daily():
    # 定义一年中每月的天数
    days_per_month = [30] * 12
    
    # 定义一年中每月的平均气温，单位：摄氏度
    monthly_avg_temps = [5, 6, 10, 15, 20, 25, 28, 27, 23, 16, 10, 6]
    
    # 定义一年中每月的降水量，单位：毫米
    monthly_precipitations = [20, 20, 30, 50, 75, 90, 110, 90, 60, 50, 30, 20]
    
    # 定义一年中每月的白天平均温度和夜晚平均温度之差，单位：摄氏度
    monthly_day_night_temperature_diffs = [15, 13, 11, 9, 7, 5, 5, 6, 7, 9, 12, 14]
    
    # 定义每天的平均气温，降水量和昼夜温差
    daily_avg_temps = []
    daily_precipitations = []
    daily_day_night_temperature_diffs = []
    
    # 循环遍历每个月
    for i in range(12):
        # 计算当前月份的每天平均气温，单位：摄氏度
        daily_temps = np.random.normal(loc=monthly_avg_temps[i], scale=2.0, size=days_per_month[i])
        daily_avg_temps.extend(daily_temps)
        
        # 计算当前月份的每天降水量，单位：毫米
        daily_precips = np.random.normal(loc=monthly_precipitations[i], scale=5.0, size=days_per_month[i])
        daily_precipitations.extend(daily_precips)
        
        # 计算当前月份的每天昼夜温差，单位：摄氏度
        daily_day_night_temps = np.random.normal(loc=monthly_day_night_temperature_diffs[i], scale=1.0, size=days_per_month[i])
        daily_day_night_temperature_diffs.extend(daily_day_night_temps)
        
    return daily_avg_temps, daily_precipitations, daily_day_night_temperature_diffs
```

```python
import matplotlib.pyplot as plt

def plot_climate_simulator_daily(year_temp, year_humidity, year_diurnal):
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
    days = range(1, 361)

    # Plot temperature
    axs[0, 0].plot(days, year_temp, marker='o')
    axs[0, 0].set_title('Daily Temperature (Celsius)')
    axs[0, 0].set_xlabel('Day')
    axs[0, 0].set_xticks(days)

    # Plot humidity
    axs[0, 1].plot(days, year_humidity, marker='o')
    axs[0, 1].set_title('Daily Humidity (%)')
    axs[0, 1].set_xlabel('Day')
    axs[0, 1].set_xticks(days)

    # Plot diurnal amplitude
    axs[1, 0].plot(days, year_diurnal, marker='o')
    axs[1, 0].set_title('diurnal amplitude (Celsius)')
    axs[1, 0].set_xlabel('Day')
    axs[1, 0].set_xticks(days)

    # Adjust spacing and show plot
    fig.tight_layout()
    plt.show()
```

```python
year_temp, year_humidity, year_diurnal = simulate_tropical_savanna_weather_daily()

plot_climate_simulator_daily(year_temp, year_humidity, year_diurnal)
```

```python
year_temp, year_humidity, year_diurnal = simulate_tropical_dessert_weather_daily()

plot_climate_simulator_daily(year_temp, year_humidity, year_diurnal)
```
