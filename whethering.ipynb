```python
import numpy as np

def simulate_tropical_savanna_weather():


    # 平均气温
    temp = []
    for i in range(12):
        # 亚热带稀树草原的平均温度为18-22摄氏度
        temp_month = np.random.normal(loc=20, scale=8)
        if i in [0,1,11]: # 冬季
            temp_month = np.random.normal(loc=10, scale=8)
        elif i in [6,7,8]: # 夏季
            temp_month = np.random.normal(loc=26, scale=8)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 亚热带稀树草原的平均湿度为50-70%
        humidity_month = np.random.normal(loc=60, scale=10)
        if i in [5,6,7]: # 雨季
            humidity_month = np.random.normal(loc=80, scale=10)
        elif i in [1,2,11]: # 旱季
            humidity_month = np.random.normal(loc=40, scale=10)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 亚热带稀树草原的平均降水量为100毫米
        precipitation_month = np.random.normal(loc=100, scale=50)
        if i in [5,6,7]: # 雨季
            precipitation_month = np.random.normal(loc=200, scale=100)
        elif i in [1,2,11]: # 旱季
            precipitation_month = np.random.normal(loc=50, scale=20)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 亚热带稀树草原的平均日照时数为2200-2600小时
        sunshine_month = np.random.normal(loc=2400, scale=200)
        if i in [0,1,11]: # 冬季
            sunshine_month = np.random.normal(loc=2000, scale=200)
        elif i in [6,7,8]: # 夏季
            sunshine_month = np.random.normal(loc=2800, scale=200)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
def simulate_tropical_rainforest_weather():

    # 平均气温
    temp = []
    for i in range(12):
        # 热带雨林气候平均温度是25-27摄氏度
        temp_month = np.random.normal(loc=25, scale=2)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 热带雨林气候的平均湿度可达80%至90%
        humidity_month = np.random.normal(loc=85, scale=8)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 热带雨林气候的平均降水量为200毫米
        precipitation_month = np.random.normal(loc=200, scale=50)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 热带雨林气候的平均光照量为1500小时
        sunshine_month = np.random.normal(loc=1500, scale=200)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
def simulate_tropical_dessert_weather():

    # 平均气温
    temp = []
    for i in range(12):
        # 热带沙漠气候平均温度是20-30摄氏度
        temp_month = np.random.normal(loc=25, scale=5)
        temp.append(temp_month)
    
    # 湿度
    humidity = []
    for i in range(12):
        # 热带沙漠气候的平均湿度可达10%至25%
        humidity_month = np.random.normal(loc=15, scale=8)
        humidity.append(humidity_month)

    # 降水量
    precipitation = []
    for i in range(12):
        # 热带沙漠气候的平均降水量5毫米
        precipitation_month = np.random.normal(loc=5, scale=3)
        precipitation.append(precipitation_month)

    # 光照
    sunshine = []
    for i in range(12):
        # 热带沙漠气候的平均光照量为2700小时
        sunshine_month = np.random.normal(loc=2700, scale=400)
        sunshine.append(sunshine_month)

    return temp, humidity, precipitation, sunshine
```

```python
import matplotlib.pyplot as plt

def plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight):
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
    months = range(1, 13)

    # Plot temperature
    axs[0, 0].plot(months, year_temp, marker='o')
    axs[0, 0].set_title('Average Monthly Temperature (Celsius)')
    axs[0, 0].set_xlabel('Month')
    axs[0, 0].set_xticks(months)

    # Plot humidity
    axs[0, 1].plot(months, year_humidity, marker='o')
    axs[0, 1].set_title('Average Monthly Humidity (%)')
    axs[0, 1].set_xlabel('Month')
    axs[0, 1].set_xticks(months)

    # Plot rainfall
    axs[1, 0].plot(months, year_rainfall, marker='o')
    axs[1, 0].set_title('Average Monthly Rainfall (mm)')
    axs[1, 0].set_xlabel('Month')
    axs[1, 0].set_xticks(months)

    # Plot sunlight
    axs[1, 1].plot(months, year_sunlight, marker='o')
    axs[1, 1].set_title('Average Monthly Sunlight (hours)')
    axs[1, 1].set_xlabel('Month')
    axs[1, 1].set_xticks(months)

    # Adjust spacing and show plot
    fig.tight_layout()
    plt.show()
```

```python
year_temp, year_humidity, year_rainfall, year_sunlight = simulate_tropical_savanna_weather()

plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight)
```

```python
year_temp, year_humidity, year_rainfall, year_sunlight = simulate_tropical_dessert_weather()

plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight)
```

```python

year_temp, year_humidity, year_rainfall, year_sunlight = simulate_tropical_rainforest_weather()

plot_climate_simulator(year_temp, year_humidity, year_rainfall, year_sunlight)
```
